```python
## training the data
x_train=np.array([1,2])
y_train=np.array([300,500])
```

```python
plt.scatter(x_train,y_train,marker='*',color='red')
plt.title('house price')
plt.xlabel('area')
plt.ylabel('price')
plt.show()
```

```python
### adding features parameters to the modle

"""
our model is represented by f(x)=a*x+b 
where a& b are its parameters also called as coefficients a and b
x= training set or feature vector
Now we will pass the feature vector to fit the model

steps:
1. analyze the model
2. f(x)=a*x+b, here we want to find the value fo f(x) for every value of x given that a and b are provided
"""
```

```python
a=100
b=100
m=x_train.shape[0]
print(m)
```

```python
f_ab=np.zeros(m)  # initializing the f(x)
print(f_ab)
```

```python
# fitting the data for f(x)
```

```python
def compute(x,a,b):
    for i in range(m):
        f_ab[i]=a*x[i]+b
    return f_ab
```

```python
yhat=compute(x_train,a,b)
print(yhat)
```

```python
plt.plot(x_train,yhat,color='blue',label='predicted_value')

plt.scatter(x_train,y_train,marker='*',color='red',label='acutal value')
plt.title('house price')
plt.xlabel('area')
plt.ylabel('price')
plt.legend()
plt.show()
```

```python
"""
cost function: it is the averaged error between predicted and the actual value

J(a,b)=1/2m*{i=1 to m (Y^hat-y)^2}
"""
```

```python
## cost function
```

```python
# calculating the cost function
def cost_evaluate(x,y,a,b):
    m=x_train.shape[0]
    cost_sum=0
    for i in range(m):
        f_ab=a*x[i]+b
        cost=(f_ab-y[i])**2
        cost_sum+=cost
    total_cost=(1/2)*m*cost_sum
    return total_cost
        
```

```python
price=cost_evaluate(x_train,y_train,100,200)
print(price)
```
